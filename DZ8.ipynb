```python
import pandas as pd
import numpy as np
#столб
columns = [
    'surgery', 'age', 'hospital_number', 'rectal_temperature', 'pulse', 'respiration_rate',
    'extremities_temperature', 'peripheral_pulse', 'mucous_membranes', 'capillary_refill',
    'pain', 'peristalsis', 'abdominal_distension', 'nasogastric_tube', 'nasogastric_reflux',
    'nasogastric_reflux_ph', 'rectal_exam_feces', 'abdomen', 'packed_cell_volume',
    'total_protein', 'abdominocentesis_appearance', 'abdomcentesis_total_protein',
    'outcome', 'surgical_lesion', 'type_lesion_site', 'type_lesion_type',
    'type_lesion_subtype', 'type_lesion_specific', 'cp_data'
]


df = pd.read_csv('horse_data.csv', header=None, names=columns, na_values='?')


numerical_cols = ['rectal_temperature', 'pulse', 'respiration_rate', 'packed_cell_volume']
categorical_cols = ['surgery', 'outcome', 'abdominal_distension', 'mucous_membranes']

selected_cols = numerical_cols + categorical_cols
df_selected = df[selected_cols]


print("Базовые метрики числовых столбцов")
for col in numerical_cols:
    print(f"\nСтолбец: {col}")
    print("Медиана:", df_selected[col].median())
    print("Среднее:", df_selected[col].mean())
    print("Мода:", df_selected[col].mode().values)
    print("Стандартное отклонение:", df_selected[col].std())
    print("Минимум:", df_selected[col].min())
    print("Максимум:", df_selected[col].max())

print("\nРаспределение категориальных столбцов")
for col in categorical_cols:
    print(f"\nСтолбец: {col}")
    print(df_selected[col].value_counts(dropna=False))


print("\n Выбросы по числовым столбцам")
for col in numerical_cols:
    Q1 = df_selected[col].quantile(0.25)
    Q3 = df_selected[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df_selected[(df_selected[col] < Q1 - 1.5*IQR) | (df_selected[col] > Q3 + 1.5*IQR)]
    print(f"\nСтолбец: {col}, количество выбросов: {len(outliers)}")
    print(outliers[[col]])


print("\n Пропуски в выбранных столбцах")
print(df_selected.isnull().sum())


df_clean = df_selected.dropna()
print("\n Размер датафрейма после удаления пропусков:", df_clean.shape)
```

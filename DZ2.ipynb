```python
number = 123321

num_str = str(number)

sum_first = int(num_str[0]) + int(num_str[1]) + int(num_str[2])
sum_last = int(num_str[3]) + int(num_str[4]) + int(num_str[5])

# Проверяем счастливость билета
if sum_first == sum_last:
    print("Счастливый билет")
else:
    print("Несчастливый билет")
```

```python
year = 2024


if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print("Високосный год")
else:
    print("Обычный год")
```

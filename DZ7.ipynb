```python
import pandas as pd
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt

df = pd.read_csv('web_clients_correct_доп материал к тема 6, 8.csv').head(15)

plt.figure(figsize=(7,5))
sns.scatterplot(data=df, x='age', y='bill', s=80)
plt.show()
```

```python
fig = px.scatter(df,
                 x='age',
                 y='bill',
                 size='bill')
fig.show()
```

```python

plt.figure(figsize=(7,5))
sns.regplot(data=df, x='age', y='bill')
plt.show()
```

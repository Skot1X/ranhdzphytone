```python
import random

def play_slot():
    number = random.randint(0, 999)
    num_str = f"{number:03d}"

    if num_str == "777":
        return 200
    elif num_str == "999":
        return 100
    elif num_str == "555":
        return 50
    elif num_str == "333":
        return 15
    elif num_str == "111":
        return 10
    elif num_str[1:] == "77":
        return 5
    elif num_str[2] == "7":
        return 3
    elif num_str[1:] == "00":
        return 2
    elif num_str[2] == "0":
        return 1
    else:
        return 0

N = 1_000_000
total_gain = 0

for _ in range(N):
    total_gain += play_slot() - 1

average_gain = total_gain / N
print(f"Средний выигрыш (проигрыш) на одну игру: {average_gain:.2f} руб")
if average_gain > 0:
    print("Игра выгодна игроку.")
else:
    print("Игра не выгодна игроку.")
```

```python
prob_777 = 1 / 1000
prob_999 = 1 / 1000
prob_555 = 1 / 1000
prob_333 = 1 / 1000
prob_111 = 1 / 1000
prob_star77 = 9 / 1000
prob_starstar7 = 90 / 1000
prob_star00 = 9 / 1000
prob_starstar0 = 90 / 1000


win_777 = 200
win_999 = 100
win_555 = 50
win_333 = 15
win_111 = 10
win_star77 = 5
win_starstar7 = 3
win_star00 = 2
win_starstar0 = 1

expected_gain = (
    prob_777 * win_777 +
    prob_999 * win_999 +
    prob_555 * win_555 +
    prob_333 * win_333 +
    prob_111 * win_111 +
    prob_star77 * win_star77 +
    prob_starstar7 * win_starstar7 +
    prob_star00 * win_star00 +
    prob_starstar0 * win_starstar0
) - 1

print(f"Теоретический средний выигрыш (проигрыш): {expected_gain:.2f} руб")
if expected_gain > 0:
    print("Игра выгодна игроку (теория).")
else:
    print("Игра не выгодна игроку (теория).")
```

```python
import csv
import json
purchase_dict = {}

with open('purchase_log.txt', 'r', encoding='utf-8') as f:
    for line in f:
        data = json.loads(line.strip())
        purchase_dict[data['user_id']] = data['category']

with open('visit_log.csv', 'r', encoding='utf-8') as f_in, \
     open('funnel.csv', 'w', encoding='utf-8', newline='') as f_out:

    reader = csv.reader(f_in)
    writer = csv.writer(f_out)

    writer.writerow(['user_id', 'source', 'category'])

    next(reader)

    for row in reader:
        user_id, source = row[0], row[1]


        if user_id in purchase_dict:
            writer.writerow([user_id, source, purchase_dict[user_id]])
```

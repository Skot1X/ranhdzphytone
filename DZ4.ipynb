```python

documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}


def get_owner_by_number(number: str) -> str:
    for doc in documents:
        if doc["number"] == number:
            return doc["name"]
    return None


def get_shelf_by_number(number: str) -> str:

    for shelf, docs in directories.items():
        if number in docs:
            return shelf
    return None


def process_command_p():
    number = input("Введите номер документа:\n")
    owner = get_owner_by_number(number)
    if owner:
        print("Владелец документа:", owner)
    else:
        print("Владелец документа: владелец не найден")


def process_command_s():
    number = input("Введите номер документа:\n")
    shelf = get_shelf_by_number(number)
    if shelf:
        print("Документ хранится на полке:", shelf)
    else:
        print("Документ не найден.")



def main():
    while True:
        command = input("Введите команду:\n")

        if command == "q":
            print("Программа завершена.")
            break

        elif command == "p":
            process_command_p()

        elif command == "s":
            process_command_s()

        else:
            print("Неизвестная команда. Доступные команды: p, s, q.")


main()
```
